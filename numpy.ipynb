```python
import numpy as np
```

```python
# Python list
list = [1,2,3,4,5]
print(list)
print(type(list))
```

```python
# NumPy array
arr = np.array([1,2,3,4,5])
print(list)
print(type(arr))
```

```python
# Creating a rank 1 Array
arr = np.array([1, 2, 3],4,5)
print(arr)
```

```python
# Creating a rank 2 Array
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print(arr)
```

```python
# Creating an array from tuple
arr = np.array((1, 3, 2))
print(arr)
```

```python
A = np.array([[1,2,3],[4,5,6]])
print(A.ndim)   # 2  (axes)
print(A.shape)  # (2, 3) dimension
print(A.size)   # 6  (total elements)
print(A.dtype)  # dtype('int64') (depends on system)
print(A.itemsize) # bytes per element
```

```python
x = [j**4 for j in range(1,9)]
print(x)

%timeit x  #how much time take to excute x
```

```python
x = np.arange(1, 9) ** 4
print(x)
%timeit x  #how much time take to excute x
```
