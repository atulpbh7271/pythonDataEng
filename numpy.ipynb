```python
import numpy as np
```

```python
# Python list
list = [1,2,3,4,5]
print(list)
print(type(list))
```

```python
# NumPy array
arr = np.array([1,2,3,4,5])
print(list)
print(type(arr))
```

```python
# Creating a rank 2 Array
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print(arr)
```
