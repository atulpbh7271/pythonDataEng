```python
import numpy as np
```

```python
# Python list
list = [1,2,3,4,5]
print(list)
print(type(list))
```

```python
# NumPy array
arr = np.array([1,2,3,4,5])
print(list)
print(type(arr))
```

```python
# Creating a rank 1 Array
arr = np.array([1, 2, 3,4,5])
print(arr)
```

```python
# Creating a rank 2 Array
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print(arr)
```

```python
# Creating an array from tuple
arr = np.array((1, 3, 2))
print(arr)
```

```python
A = np.array([[1,2,3],[4,5,6]])
print(A.ndim)   # 2  (axes)
print(A.shape)  # (2, 3) dimension
print(A.size)   # 6  (total elements)
print(A.dtype)  # dtype('int64') (depends on system)
print(A.itemsize) # bytes per element
```

```python
x = [j**4 for j in range(1,9)]
print(x)

%timeit x  #how much time take to excute x
```

```python
x = np.arange(1, 9) ** 4
print(x)
%timeit x  #how much time take to excute x
```

```python
#Dimensions in Arrays

#0-D Arrays
arr = np.array(42)

print(arr)
```

```python
#1-D Arrays

arr = np.array([1, 2, 3, 4, 5])

print(arr)
```

```python
#2-D Arrays
arr = np.array([[1, 2, 3], [4, 5, 6]])

print(arr)
```

```python
#3-D arrays
arr = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])

print(arr)
```

```python
#Check Number of Dimensions
x = np.array(42)
y = np.array([1, 2, 3, 4, 5])
z = np.array([[1, 2, 3], [4, 5, 6]])
a = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])

print(x.ndim)
print(y.ndim)
print(z.ndim)
print(a.ndim)
```

```python
#Higher Dimensional Arrays
# ndim used for any number of dimensions

arr = np.array([1, 2, 3, 4], ndmin=5)

print(arr)
print('number of dimensions :', arr.ndim)
```

```python
x=np.array([1,2,3])                 # from list
y=np.array([[1,2],[3,4]], dtype=float)
z=np.asarray([1,2,3])               # like array, avoids copy if already ndarray

print(x)
print(y)
print(z)
```

```python
#Ranges & evenly spaced
x=np.arange(0, 10, 2)               # [0 2 4 6 8]
y=np.linspace(0, 1, 5)              # [0. , 0.25, 0.5 , 0.75, 1. ]

# Log-spaced
z=np.logspace(1, 3, 3)              # [   10.   100.  1000.]

print(x)
print(y)
print(z)
```

```python
#Filled arrays
x=np.zeros((2,3))                   # 2x3 zeros
y=np.ones((3,2), dtype=int)         # ones
z=np.full((2,2), fill_value=7.5)    # constant
a=np.empty((2,3))                   # uninitialized (whatever is in memory)

print(x)
print(y)
print(z)
print(a)

```
